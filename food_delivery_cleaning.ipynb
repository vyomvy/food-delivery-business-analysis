```
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/food_delivery_data.csv')
```

```
df.shape
```

```
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
)
```

```
df.info()
```

```
df['order_date'] = df['order_date'].apply(pd.to_datetime)
```

```
df.duplicated().sum()
```

```
df.drop_duplicates(inplace=True)
```

```
df['distance_km'].describe()
```

```
df['payment_method'] = df['payment_method'].str.replace('null', 'unknown', case=False)
```

```
df['area'].unique()
```

```
df['area'] = (
      df['area']
      .str.strip()
      .str.lower()
      .str.replace(' ', '')
)
```

```
df['food_item'] = (
      df['food_item']
      .str.strip()
      .str.lower()
      .str.replace(' ', '')
)
```
